```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Read the data set of "flavors_of_cacao" in data.
data = pd.read_csv("C:/Users/Acer/OneDrive/Desktop/Python/flavors_of_cacao.csv")

#Printing the data
data
```

```python
#Dropping the column 'Bean_Type' from the data frame.
data.drop('Bean_Type', axis = 1, inplace = True)

#Printing the Dataset
data
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Broad_Bean_Origin']
```

```python
#Dropping the records with Broad Bean Origin missing in data frame.
data = data[~data.Broad_Bean_Origin.isnull()].copy()

#Cheacking the missing values in the dataset.
data.isnull().sum()
```

```python
data
```

```python
#Let's calculate the rating category
data.Rating.value_counts(normalize=True)

#Plot the bar graph of Rating categories
data.Rating.value_counts(normalize=True).plot.barh()
plt.show()
```

```python
#calculate the percentage of each Broad_Bean_Origin category.
data.Broad_Bean_Origin.value_counts(normalize=True)

#plot the pie chart of Broad_Bean_Origin categories
data.Broad_Bean_Origin.value_counts(normalize=True).plot.pie()
plt.show()
```

```python
data.Rating.describe()
```

```python
#plot the scatter plot of REF and Rating variable in data
plt.scatter(data.REF,data.Rating)
plt.show()

#plot the scatter plot of REF in x axis and Rating in y axis
data.plot.scatter(x="REF",y="Rating")
plt.show()
```

```python
#Plot the pair plot of REF, Rating and Review Date in data dataframe.
sns.pairplot(data=data, vars=['REF','Rating','Review_Date'])
plt.show()
```

```python
#creating a matrix using REF, Rating and Review Date as rows and columns
data[['REF','Rating','Review_Date']].corr()

sns.heatmap(data[['REF','Rating','Review_Date']].corr(), annot=True, cmap='Reds')
plt.show()
```

```python
#plot the box plot of Rating for Broad_Bean_Origin
sns.boxplot(data.Review_Date, data.Rating)
plt.show()
```
